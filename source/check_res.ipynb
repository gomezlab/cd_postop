```python
import pandas as pd
import os
```

```python
res_dfs = {}
for t_type in ['c', 'i']:
    if t_type == 'c':
        res_dfs[t_type] = pd.read_csv(f'../results/recur/{t_type}/ruvg_runs/res5_19_symbol.csv', index_col=0)
    else:
        res_dfs[t_type] = pd.read_csv(f'../results/recur/{t_type}/ruvg_runs/res5_18_symbol.csv', index_col=0)
```

```python
res_dfs['c'].head()
```

```python
gene_list = ['AQP9', 'HCAR3']
select_gene_df = pd.DataFrame(columns=['gene', 'tissue', 'log2FoldChange', 'padj'])
for gene in gene_list:
    for tissue in ['c', 'i']:
        gene_res = res_dfs[tissue].loc[res_dfs[tissue].index == gene]
        print(gene, tissue)
        gene_res['gene'] = gene
        gene_res['tissue'] = tissue
        gene_res = gene_res[['gene', 'tissue', 'log2FoldChange', 'padj']]
        select_gene_df = pd.concat([select_gene_df, gene_res], axis=0)
select_gene_df = select_gene_df.reset_index(drop=True)
select_gene_df.head()
```
