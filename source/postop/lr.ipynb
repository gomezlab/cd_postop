```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
```

```python
data = pd.read_csv('../../data/coldata_nsqip_clean.csv', index_col=0)
data.head()
```

```python
path_data = pd.read_csv('results/hall_paths.csv', index_col=0)
path_data.head()
```

```python
path_data = path_data.T
path_data.head()
```

```python
allo_cols = [col for col in path_data.columns if 'ALLO' in col]
allo_cols
```

```python
data['jak_stat'] = path_data['HALLMARK_IL6_JAK_STAT3_SIGNALING']
data['tnf_alpha'] = path_data['HALLMARK_TNFA_SIGNALING_VIA_NFKB']
data['il2_stat5'] = path_data['HALLMARK_IL2_STAT5_SIGNALING']
data['allo_reject'] = path_data['HALLMARK_ALLOGRAFT_REJECTION']
data['ox_phos'] = path_data['HALLMARK_OXIDATIVE_PHOSPHORYLATION']
data.head()
```

```python
#for each of the pathway columns, quantile bin the values
for col in ['jak_stat', 'tnf_alpha', 'il2_stat5', 'allo_reject', 'ox_phos']:
    data[str(col)+'_bin'] = data[col].apply(lambda x: '1' if x > data[col].quantile(0.5) else '0')
```

```python
cat_cols = ['Sex', 'Race', 'ASA Classification', 'Smoking Status', 'Diabetes', 'Antibiotic Bowel Prep']
num_cols = ['Age', 'Hematocrit', 'tnf_alpha', 'jak_stat', 'il2_stat5', 'allo_reject', 'ox_phos']
y = ['Surgical Site Infection']
data_ssi = data[cat_cols + num_cols + y]
data_ssi.columns = [col.replace(' ', '_') for col in data_ssi.columns]
cat_cols = [col.replace(' ', '_') for col in cat_cols]
data_ssi.head()
```

```python
from tableone import TableOne
y = 'Surgical_Site_Infection'
tab1 = TableOne(data_ssi, columns=cat_cols + num_cols, categorical=cat_cols, groupby=y, pval=True)
tab1
```

```python
pval_df = pd.DataFrame(columns=['pathway', 'pval'])
for pathway in ['tnf_alpha', 'jak_stat', 'il2_stat5', 'allo_reject', 'ox_phos']:
    try:
        formula = "Surgical_Site_Infection ~ C(Race, Treatment(reference='Non-Hispanic White')) + C(Diabetes, Treatment(reference='No diabetes')) + Hematocrit + {}".format(pathway)
        model = sm.formula.glm(formula=formula, data=data_ssi, family=sm.families.Binomial()).fit()
        pval = model.pvalues[pathway]
        pval_df = pd.concat([pval_df, pd.DataFrame({'pathway': [pathway], 'pval': [pval]})])
    except:
        print(pathway)
        print('error')
        pass
pval_df
```

```python
cat_cols = ['Sex', 'Race', 'ASA Classification', 'Smoking Status', 'Diabetes', 'Antibiotic Bowel Prep', 'tnf_alpha_bin', 'jak_stat_bin', 'il2_stat5_bin', 'allo_reject_bin', 'ox_phos_bin']
num_cols = ['Age', 'Hematocrit']
y = ['Surgical Site Infection']
data_ssi = data[cat_cols + num_cols + y]
data_ssi.columns = [col.replace(' ', '_') for col in data_ssi.columns]
cat_cols = [col.replace(' ', '_') for col in cat_cols]
data_ssi['tnf_alpha_bin'].value_counts()
```

```python

#now do it again with the binarized pathways
pval_df = pd.DataFrame(columns=['pathway', 'pval'])
pathway = 'tnf_alpha_bin'

formula = "Surgical_Site_Infection ~ C(Race, Treatment(reference='Non-Hispanic White')) + C(Diabetes, Treatment(reference='No diabetes')) + C({}) + Hematocrit".format(pathway)
model = sm.formula.glm(formula=formula, data=data_ssi, family=sm.families.Binomial()).fit()
model.pvalues
```

```python

pval = model.pvalues['C({})[T.1]'.format(pathway)]
pval_df = pd.concat([pval_df, pd.DataFrame({'pathway': [pathway], 'pval': [pval]})])
pval_df
```

```python
#now do it again with the binarized pathways
pval_df = pd.DataFrame(columns=['pathway', 'pval'])
pathway = ['tnf_alpha_bin']
for pathway in ['tnf_alpha_bin', 'jak_stat_bin', 'il2_stat5_bin', 'allo_reject_bin', 'ox_phos_bin']:
    try:
        formula = "Surgical_Site_Infection ~ C(Race, Treatment(reference='Non-Hispanic White')) + C(Diabetes, Treatment(reference='No diabetes')) + C({}) + Hematocrit".format(pathway)
        model = sm.formula.glm(formula=formula, data=data_ssi, family=sm.families.Binomial()).fit()
        pval = model.pvalues['C({})[T.1]'.format(pathway)]
        pval_df = pd.concat([pval_df, pd.DataFrame({'pathway': [pathway], 'pval': [pval]})])
    except:
        print(pathway)
        #print the error that occurs
        print('error')
        
        pass
pval_df
```

```python
formula = "Surgical_Site_Infection ~ C(Race, Treatment(reference='Non-Hispanic White')) + C(Diabetes, Treatment(reference='No diabetes')) + Hematocrit + C(ox_phos_bin)"
model = sm.formula.glm(formula=formula, data=data_ssi, family=sm.families.Binomial()).fit()
model.summary()
```

```python
#save the results of the oxidative phosphorylation model to csv
res_df = model.summary2().tables[1]
res_df.head()
```
