```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('coldata_nsqip.csv', index_col=0)
data.head()
```

```python
pd.crosstab(data['IASC'], data['SSI'], margins=True)
```

```python
data['SSI_IASC'] = np.NaN
data.loc[data['SSI'] == 'No', 'SSI_IASC'] = 'No infection'
data.loc[data['SSI'] == 'Yes', 'SSI_IASC'] = 'SSI'
data.loc[data['IASC'] == 'Yes', 'SSI_IASC'] = 'IASC'
data['SSI_IASC'].value_counts(dropna=False)
```

```python
data.to_csv('coldata_nsqip.csv')
```
