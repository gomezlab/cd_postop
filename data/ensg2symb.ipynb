```python
import pandas as pd
import os
```

```python
cts_c = pd.read_csv('cts_c.csv', index_col=0)
cts_c.head()
```

```python
cts_i = pd.read_csv('cts_i.csv', index_col=0)
cts_i.head()
```

```python
ensg2gene = pd.read_csv('hg38_ensembl2gene.txt', sep='\t', index_col=0)
ensg2gene['gene_id'] = ensg2gene['gene_id'].str.lstrip(' ')
ensg2gene.head()
```

```python
cts_c['symbol'] = cts_c.index.map(ensg2gene['gene_id'])
cts_c.set_index('symbol', inplace=True)
cts_c.head()
```

```python
#drop duplicates in index
print(cts_c.shape)
cts_c = cts_c[~cts_c.index.duplicated(keep='first')]
print(cts_c.shape)
cts_c.head()
```

```python
cts_i['symbol'] = cts_i.index.map(ensg2gene['gene_id'])
cts_i.set_index('symbol', inplace=True)
cts_i.head()
```

```python
print(cts_i.shape)
cts_i = cts_i[~cts_i.index.duplicated(keep='first')]
print(cts_i.shape)
cts_i.head()
```

```python
cts_c.to_csv('cts_c_symbol.csv')
cts_i.to_csv('cts_i_symbol.csv')
```
